```python
import pandas as pd
import sqlite3

# --- 1. ЗАГРУЗКА И ПОДГОТОВКА ДАННЫХ ---
# Правильный путь к файлу в Google Colab
csv_file_path = '/content/data.csv'

try:
    df = pd.read_csv(csv_file_path, encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv(csv_file_path, encoding='latin1')

# Преобразование типов данных
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df['CustomerID'] = pd.to_numeric(df['CustomerID'], errors='coerce')
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'], errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df = df.where(pd.notna(df), None)

# Создание столбца 'total_price'
df['total_price'] = df['Quantity'] * df['UnitPrice']

# Создание базы данных в памяти
conn = sqlite3.connect(':memory:')
df.to_sql('sales', conn, index=False)
print("Данные успешно загружены в базу данных SQLite в памяти.")

# --- 2. ВЫПОЛНЕНИЕ SQL-ЗАПРОСОВ ---

# Запрос 1: Общая выручка за период
query_1 = """
SELECT SUM(total_price) AS total_revenue
FROM sales;
"""
result_1 = pd.read_sql_query(query_1, conn)
print("\n--- 1. Общая выручка за период ---")
print(result_1)

# Запрос 2: Количество уникальных клиентов
query_2 = """
SELECT COUNT(DISTINCT CustomerID) AS total_unique_customers
FROM sales;
"""
result_2 = pd.read_sql_query(query_2, conn)
print("\n--- 2. Количество уникальных клиентов ---")
print(result_2)

# Запрос 3: Средний чек
query_3 = """
SELECT AVG(total_price) AS average_order_value
FROM sales;
"""
result_3 = pd.read_sql_query(query_3, conn)
print("\n--- 3. Средний чек ---")
print(result_3)

# Запрос 4: Самые популярные товары
query_4 = """
SELECT Description, SUM(Quantity) AS total_sold
FROM sales
GROUP BY Description
ORDER BY total_sold DESC
LIMIT 10;
"""
result_4 = pd.read_sql_query(query_4, conn)
print("\n--- 4. Самые популярные товары ---")
print(result_4)

# Запрос 5: Продажи по странам
query_5 = """
SELECT Country, SUM(total_price) AS total_revenue
FROM sales
GROUP BY Country
ORDER BY total_revenue DESC
LIMIT 10;
"""
result_5 = pd.read_sql_query(query_5, conn)
print("\n--- 5. Продажи по странам ---")
print(result_5)

# Запрос 6: Продажи по месяцам (trend)
query_6 = """
SELECT
    strftime('%Y-%m', InvoiceDate) AS sales_month,
    SUM(total_price) AS total_revenue
FROM sales
GROUP BY sales_month
ORDER BY sales_month;
"""
result_6 = pd.read_sql_query(query_6, conn)
print("\n--- 6. Продажи по месяцам ---")
print(result_6)


# Закрываем соединение
conn.close()
```

```python
import matplotlib.pyplot as plt

# График 1: Тенденция продаж по месяцам
plt.figure(figsize=(12, 6))
plt.plot(result_6['sales_month'], result_6['total_revenue'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# График 2: Топ-10 самых популярных товаров
plt.figure(figsize=(12, 6))
plt.bar(result_4['Description'], result_4['total_sold'], color='skyblue')
plt.title('Top 10 Most Popular Products')
plt.xlabel('Product Description')
plt.ylabel('Total Sold')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# График 3: Продажи по странам
plt.figure(figsize=(12, 6))
plt.bar(result_5['Country'], result_5['total_revenue'], color='lightgreen')
plt.title('Total Revenue by Country')
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
